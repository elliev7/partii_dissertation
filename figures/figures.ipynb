```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#collecting data (in ms)
#pinging for 100 samples

# Router4 IPv4 
#Host 1 to Host 2
r4_ip4_hh_min    = 1.412
r4_ip4_hh_max    = 2.819 
r4_ip4_hh_mean   = 1.530
r4_ip4_hh_stddev = 0.145

#Host 1 to Router
r4_ip4_hr_min    = 0.858
r4_ip4_hr_max    = 1.740
r4_ip4_hr_mean   = 0.962
r4_ip4_hr_stddev = 0.147


#Router6 IPv6
#Host 1 to Host 2
r6_ip6_hh_min    = 1.280
r6_ip6_hh_max    = 3.292
r6_ip6_hh_mean   = 1.574
r6_ip6_hh_stddev = 0.239

#Host 1 to Router
r6_ip6_hr_min    = 0.747
r6_ip6_hr_max    = 1.904
r6_ip6_hr_mean   = 0.946
r6_ip6_hr_stddev = 0.116


#DualStack IPv4
#Host 1 to Host 2
ds_ip4_hh_min    = 1.459
ds_ip4_hh_max    = 2.936
ds_ip4_hh_mean   = 1.656
ds_ip4_hh_stddev = 0.139

#Host 1 to Router
ds_ip4_hr_min    = 0.866
ds_ip4_hr_max    = 1.657
ds_ip4_hr_mean   = 1.077
ds_ip4_hr_stddev = 0.091


#DualStack IPv6
#Host 1 to Host 2
ds_ip6_hh_min    = 1.422
ds_ip6_hh_max    = 3.340
ds_ip6_hh_mean   = 1.716
ds_ip6_hh_stddev = 0.288

#Host 1 to Router
ds_ip6_hr_min    = 0.883
ds_ip6_hr_max    = 1.924
ds_ip6_hr_mean   = 1.062
ds_ip6_hr_stddev = 0.100
```

```python
x_r4_ip4_hh = np.linspace(r4_ip4_hh_mean - 3*r4_ip4_hh_stddev, r4_ip4_hh_mean + 3*r4_ip4_hh_stddev, 100)
y_r4_ip4_hh = norm.pdf(x_r4_ip4_hh, r4_ip4_hh_mean, r4_ip4_hh_stddev)

x_r6_ip6_hh = np.linspace(r6_ip6_hh_mean - 3*r6_ip6_hh_stddev, r6_ip6_hh_mean + 3*r6_ip6_hh_stddev, 100)
y_r6_ip6_hh = norm.pdf(x_r6_ip6_hh, r6_ip6_hh_mean, r6_ip6_hh_stddev)

x_ds_ip4_hh = np.linspace(ds_ip4_hh_mean - 3*ds_ip4_hh_stddev, ds_ip4_hh_mean + 3*ds_ip4_hh_stddev, 100)
y_ds_ip4_hh = norm.pdf(x_ds_ip4_hh, ds_ip4_hh_mean, ds_ip4_hh_stddev)

x_ds_ip6_hh = np.linspace(ds_ip6_hh_mean - 3*ds_ip6_hh_stddev, ds_ip6_hh_mean + 3*ds_ip6_hh_stddev, 100)
y_ds_ip6_hh = norm.pdf(x_ds_ip6_hh, ds_ip6_hh_mean, ds_ip6_hh_stddev)

plt.figure(figsize=(10, 5))

plt.xlim(0.5, 3)
plt.ylim(0, 3)

plt.plot(x_r4_ip4_hh, y_r4_ip4_hh, linewidth=2, color='green', label ="Router4 IPv4")
plt.plot(x_r6_ip6_hh, y_r6_ip6_hh, linewidth=2, color='blue', label ="Router6 IPv6")
plt.plot(x_ds_ip4_hh, y_ds_ip4_hh, linewidth=2, color='black', label ="DualStack IPv4")
plt.plot(x_ds_ip6_hh, y_ds_ip6_hh, linewidth=2, color='red', label ="DualStack IPv6")

plt.xlabel('Time (ms)')
plt.ylabel('Probability Density')
plt.legend()

plt.savefig('Ping_HH.png')
plt.show()
```

```python
x_r4_ip4_hr = np.linspace(r4_ip4_hr_mean - 3*r4_ip4_hr_stddev, r4_ip4_hr_mean + 3*r4_ip4_hr_stddev, 100)
y_r4_ip4_hr = norm.pdf(x_r4_ip4_hr, r4_ip4_hr_mean, r4_ip4_hr_stddev)

x_r6_ip6_hr = np.linspace(r6_ip6_hr_mean - 3*r6_ip6_hr_stddev, r6_ip6_hr_mean + 3*r6_ip6_hr_stddev, 100)
y_r6_ip6_hr = norm.pdf(x_r6_ip6_hr, r6_ip6_hr_mean, r6_ip6_hr_stddev)

x_ds_ip4_hr = np.linspace(ds_ip4_hr_mean - 3*ds_ip4_hr_stddev, ds_ip4_hr_mean + 3*ds_ip4_hr_stddev, 100)
y_ds_ip4_hr = norm.pdf(x_ds_ip4_hr, ds_ip4_hr_mean, ds_ip4_hr_stddev)

x_ds_ip6_hr = np.linspace(ds_ip6_hr_mean - 3*ds_ip6_hr_stddev, ds_ip6_hr_mean + 3*ds_ip6_hr_stddev, 100)
y_ds_ip6_hr = norm.pdf(x_ds_ip6_hr, ds_ip6_hr_mean, ds_ip6_hr_stddev)

plt.figure(figsize=(10, 5))

plt.plot(x_r4_ip4_hr, y_r4_ip4_hr, linewidth=2, color='green', label ="Router4 IPv4")
plt.plot(x_r6_ip6_hr, y_r6_ip6_hr, linewidth=2, color='blue', label ="Router6 IPv6")
plt.plot(x_ds_ip4_hr, y_ds_ip4_hr, linewidth=2, color='black', label ="DualStack IPv4")
plt.plot(x_ds_ip6_hr, y_ds_ip6_hr, linewidth=2, color='red', label ="DualStack IPv6")

plt.xlim(0.5, 1.5)
plt.ylim(0, 4.5)

plt.xlabel('Time (ms)')
plt.ylabel('Probability Density')
plt.legend()

plt.savefig('Ping_HR.png')
plt.show()
```
